```python
import ast
import pandas as pd
```

```python
steam = pd.read_csv('dataset/steam_games.csv')
```

```python
#visualizaremos el dataframe para tener una vision mas general de los datos y ver si hay datos redundantes
steam
```

```python
#A simple vista podemos observar que la columna app_name y title son iguales por lo que eliminaremos la columna app_name
steam.drop(['app_name'], axis=1, inplace=True)
steam
```

```python
#Para nuestra API, no se nos requieren datos como URL, ID, o el nombre de la compañía, por lo que podemos eliminarlos.
steam.drop(['url', 'reviews_url', 'developer',  'publisher'], axis=1, inplace=True)
steam.head()
```

```python
#Como podemos ver logramos un dataframe mucho mas optimo para las consultas que queremos realizar

    
#evaluaremos cada genero y lo guardaremos en un diccionario
specs = {}
for i in range(len(steam)):
    for j in range(len(steam[i])):
        if steam[i][j] in specs:
            specs[steam[i][j]] += 1
        else:
            specs[steam[i][j]] = 1
    
generosOrdenados = sorted(specs.items(), key=lambda x: x[1], reverse=True)
#Haremos una gradica para ver cuantos tipos de specs hay y una comparacion entre ellos
```
